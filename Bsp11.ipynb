```python
# Zeit vs Temp mit linearem fit
time = [0,5,10,15,20,25,30,35,40,45,50,55,60,65,70,75]
temp = [19.5,20.1,20.6,21.1,21.5,22.1,22.5,23,23.4,23.8,24.3,24.8,25.2,25.6,26,26.5]

import matplotlib.pyplot as plt
plt.plot(time,temp)

import numpy as np
from scipy import stats

res = stats.linregress(time[:],temp[:])
print(res)

yfit = [res.slope * i + res.intercept for i in time]
plt.scatter(time,temp, label='Temperatur')
plt.plot(time, yfit , color='red', label=f'Fit: y = {res.slope:.2f}x + {res.intercept:.2f}\R: {res.rvalue:.4f}')
plt.legend()
plt.ylabel('Temperatur [Â°C]')
plt.xlabel('Zeit [min]')
plt.show()
```
